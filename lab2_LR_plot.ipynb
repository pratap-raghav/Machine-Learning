```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def estimate_coef(x,y):
    n=np.size(x)
    mx, my = np.mean(x), np.mean(y)
    SSxy = np.sum(y*x) - n*my*mx
    SSxx = np.sum(x*x) - n*mx*mx 
    b1 = SSxy/SSxx
    b0 = my - b1*mx
    return (b0, b1)
```

```python
def plot_regression_line(x,y,b):
    plt.scatter(x,y,color = "RED",marker="o", s = 10)
    yPred = b[0] + b[1]*x
    plt.plot(x, yPred, color="b")
    plt.xlabel("x - Direction")
    plt.ylabel("y - Direction")
    plt.show()
```

```python
xl = list()
yl = list()
while(len(xl)!=50):
    _x = random.randint(0, 100)
    if _x not in xl:
        xl.append(_x)
    
while(len(yl)!=50):
    _y = random.randint(0, 10000)
    if _y not in yl:
            yl.append(_y)
    
# x = np.array([2,36,78,7,8,93,26,70,6,71,41,3,26,53,76,57,4,52,34,4])
# y = np.array([12,32,43,56,67,87,54,78,89,98,65,8,45,79,45,36,1,15,42,26])

x = np.array(xl)
y = np.array(yl)
b = estimate_coef(x,y)
print("Estimate - b0 = {} b1 = {}".format(b[0], b[1]))
plot_regression_line(x,y,b)
```
